```python
import json
import matplotlib.pyplot as plt
import numpy as np
from helpers import METRICS_JSON_PATH, PLOT_DIR, load_metrics_store
```

```python
store = load_metrics_store()
print(json.dumps(store, indent=2, ensure_ascii=False))
```

```python
names = list(store.keys())
f1 = [store[k]['metrics']['f1_macro'] for k in names]
plt.figure(figsize=(10,4))
plt.bar(names, f1)
plt.ylabel('macro F1')
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.savefig(PLOT_DIR / 'metrics_comparison.png', dpi=150)
plt.show()
plt.close()
print('Saved', PLOT_DIR / 'metrics_comparison.png')
```
